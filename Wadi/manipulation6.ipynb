```python
import pandas as pd 
df=pd.read_csv('data/ranklist.csv')
```

```python
df.columns
```

```python
df_new=df.drop(df.columns[0],axis=1)
```

```python
df_new
```

```python
score_cols={}
for i in range(97):
    for j in range(4):
        if df_new.iloc[i][j] not in score_cols:
            score_cols[df_new.iloc[i][j]]=1/(i+1)
        else:
            score_cols[df_new.iloc[i][j]]+=(1/(i+1))
```

```python
score_cols
```

```python
score_cols_sorted=dict(sorted(score_cols.items(),key=lambda x:x[1],reverse=True))
score_cols_sorted
```
