```python
import pandas as pd
import numpy as np
```

```python
train_data=np.load('traindata_ss.npy',allow_pickle=True)
```

```python
train_data.shape
```

```python
sum_mean=0
means = np.asarray([0]*124)
# print(type(means))
for hourdata in train_data:
    means = means + np.mean(hourdata,axis=0)
means = means/270
```

```python
sensor_means = []
for sensor in range(123):
  sum_p,sum_n,cnt_p,cnt_n = 0,0,0,0 
  for hour in range(270):
    for second in range(3600):
        if train_data[hour][second][-1] == 1:
          sum_p += (train_data[hour][second][sensor]-means[sensor])
          cnt_p += 1
        else:
          sum_n += (train_data[hour][second][sensor]-means[sensor])
          cnt_n += 1
  sensor_means.append([sum_p/cnt_p,sum_n/cnt_n])
sensor_means
  
```

```python
sensors_stds=[]
for sensor in range(123):
  sensorlist_n=[]
  sensorlist_a=[]
  for hour in range(270):
    for second in range(3600):
      if train_data[hour][second][-1] == 1:
        sensorlist_n.append(train_data[hour][second][sensor])
      else:
        sensorlist_a.append(train_data[hour][second][sensor])

  sensors_stds.append([np.array(sensorlist_n).std(),np.array(sensorlist_a).std()])
sensors_stds
```

```python
rank1=[]
for i in range(123):
  if sensor_means[i][0] or sensor_means[i][1]:
    temp=abs(sensor_means[i][0] - sensor_means[i][1])
    rank1.append(temp/max(sensors_stds[i][0],sensors_stds[i][1]))
  else:
    rank1.append(0.0)
```

```python
cols_to_deleted1=[i for i,x in enumerate(rank1) if x == 0.0]
print(cols_to_deleted1)
```

```python
rank1.sort(reverse=True)
rank1
```

```python
rank2=[]
cols_to_deleted2 =[]
for i in range(123):
  if sensors_stds[i][0] or sensors_stds[i][1]:
    temp=abs(sensors_stds[i][0] - sensors_stds[i][1])
    rank2.append(temp/max(sensors_stds[i][0],sensors_stds[i][1]))
  else:
    cols_to_deleted2.append(i)
    rank2.append(0.0)
print(cols_to_deleted2)
```
